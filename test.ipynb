```python
from bs4 import BeautifulSoup
import requests
import re
import pandas as pd
```

```python
url = 'https://www.albatrosmedia.eu/book/a-big-book-of-poison/'#'https://www.albatrosmedia.eu/book/a-big-book-of-opposites/' #https://www.albatrosmedia.eu/book/a-big-book-of-poison/'#  '# ##'https://www.albatrosmedia.eu/book/how-its-done-design/'#'https://www.albatrosmedia.eu/category/all/'
page = requests.get(url) 
if page.status_code == 200:
    Bs_data = BeautifulSoup(page.content, 'html.parser')
```

```python
print(Bs_data)
```

```python
age_groups_bs = Bs_data.find_all(href=re.compile("https://www.albatrosmedia.eu/category/all/"))#class_='menu-item menu-item-type-taxonomy menu-item-object-category'
```

```python
age_groups_bs
```

```python
taxonomy_bs = Bs_data.find_all(href=re.compile("https://www.albatrosmedia.eu/category/"))
```

```python
taxonomy_bs = list(filter(lambda x: x not in age_groups_bs, taxonomy_bs))  
```

```python
taxonomy_bs
```

```python

# Loop through and extract text + href
for category in taxonomy_bs:
    title = category.text.strip()  # Get category title
    url = category["href"]  # Get category URL
    print(f"Title: {title}, URL: {url}")
```

```python
url = 'https://www.albatrosmedia.eu/category/art-and-music/'
page = requests.get(url) 
if page.status_code == 200:
    Bs_data = BeautifulSoup(page.content, 'html.parser')
```

```python
book_links = Bs_data.find_all("a", class_="book-title")

# Extract and print book titles and URLs
for book in book_links:
    title = book.text.strip()  # Get book title
    url = book["href"]  # Get book URL
    print(f"Title: {title}, URL: {url}")
```

```python
podklady_albatros = pd.read_excel('podklady Albatros CRM.xlsx', sheet_name = 'knihy detail')
scraped_data = pd.read_excel('Scraped Data.xlsx')

joind_df = podklady_albatros.merge(scraped_data, left_on=['Book Name', 'Author'], right_on=['title', 'writers'], suffixes=(False, False))

collapsed_df = joind_df.groupby('Book Name', as_index=False).agg({
    'category': 'unique',  # Get all categories as a list
    'Author': 'first',  # Get the first Author for each Book Name
    'Series: Series Name': 'first',  # Get the first Series Name for each Book Name
    'Series': 'first',  # Get the first Series for each Book Name
    'PDF URL': 'first',  # Get the first PDF URL for each Book Name
    'Picture URL': 'first',  # Get the first Picture URL for each Book Name
    'YouTube URL': 'first',  # Get the first YouTube URL for each Book Name
    'Book: Created Date': 'first',  # Get the first Created Date for each Book Name
    'url': 'first',  # Get the first URL for each Book Name
    'illustrators': 'first',  # Get the first illustrator for each Book Name
    'ages': 'first',  # Get unique ages for each Book Name
    'book parametes':'first', 
    'sold to': 'first', 
    'description': 'first'
})

collapsed_df = collapsed_df.rename(columns={'url': "Albatros URL", 'illustrators': "Illustrators", 'ages': "Ages", 'book parametes': 'Book parameters', 'sold to': 'Sold to', 'description': 'Description'})
collapsed_df.to_excel('Albatros scraping.xlsx', index=False)
```
